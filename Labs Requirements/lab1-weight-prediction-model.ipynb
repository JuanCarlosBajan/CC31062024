```python
import math
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# 0. Cargar los datos de estatura y peso
url = 'http://wiki.stat.ucla.edu/socr/index.php?title=SOCR_Data_Dinov_020108_HeightsWeights&oldid=12529'

height_weight_df = pd.read_html(url)[1][['Height(Inches)','Weight(Pounds)']]
```

# 1. Contar registros en el DataFrame

```python
record_count = height_weight_df.shape[0]
print(f'Número de registros: {record_count}')
```

# 2. Crear una regresión lineal con estatura como variable explicatoria y peso como variable explicada


```python
X = height_weight_df[['Height(Inches)']]
y = height_weight_df['Weight(Pounds)']
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

# 3. Hallar el intercepto


```python
intercept = reg.intercept_
print(f'Intercepto: {intercept}')
```

# 4. Hacer la gráfica


```python
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X, reg.predict(X), color='red', label='Regresión Lineal')
plt.xlabel('Estatura (Inches)')
plt.ylabel('Peso (Pounds)')
plt.title('Regresión Lineal entre Estatura y Peso')
plt.legend()
plt.show()
```

# 5. Hallar correlación y valor p


```python
correlation, p_value = pearsonr(height_weight_df['Height(Inches)'], height_weight_df['Weight(Pounds)'])
print(f'Correlación: {correlation}')
print(f'Valor p: {p_value}')
```

# 6. Interpretar el valor p


```python
if p_value < 0.05:
    print("El valor p es menor que 0.05, lo que indica que la correlación entre estatura y peso es estadísticamente significativa.")
else:
    print("El valor p es mayor que 0.05, lo que indica que no hay suficiente evidencia para afirmar que existe una correlación significativa entre estatura y peso.")
```
